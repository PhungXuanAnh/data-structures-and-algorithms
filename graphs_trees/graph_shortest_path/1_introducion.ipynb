Hãy tưởng tượng bạn đang sử dụng Google Maps để tìm đường đi từ nhà đến một quán trà sữa. Bản đồ sẽ chỉ cho bạn thấy nhiều con đường khác nhau, nhưng mục tiêu của bạn là tìm ra con đường nhanh nhất (hoặc ngắn nhất).

Trong Cấu trúc dữ liệu và Giải thuật, đây chính là **Bài toán Đường đi ngắn nhất (Shortest Path Problem)**.

Để máy tính hiểu được điều này, chúng ta mô hình hóa bản đồ thành một **Đồ thị (Graph)** với các thành phần sau:

* **Các Đỉnh (Vertices/Nodes):** Đại diện cho các địa điểm (ví dụ: ngã tư, thành phố, hoặc ngôi nhà của bạn).
* **Các Cạnh (Edges):** Đại diện cho các con đường nối giữa các địa điểm đó.
* **Trọng số (Weights):** Là "chi phí" để đi qua con đường đó. Chi phí này có thể là độ dài (km), thời gian (phút), hoặc phí cầu đường.

Mục tiêu của bài toán là tìm một lộ trình đi từ điểm bắt đầu (A) đến điểm đích (B) sao cho **tổng trọng số** của các cạnh trên con đường đó là nhỏ nhất.
