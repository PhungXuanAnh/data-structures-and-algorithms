## 1. Code Python cho thuáº­t toÃ¡n Dijkstra

```python
import heapq  # ThÆ° viá»‡n giÃºp láº¥y giÃ¡ trá»‹ nhá» nháº¥t cá»±c nhanh

def dijkstra(graph, start):
    # ---------------------------------------------------------
    # 1. KHá»I Táº O (INITIALIZATION)
    # ---------------------------------------------------------
    
    # distances: Báº£ng lÆ°u khoáº£ng cÃ¡ch ngáº¯n nháº¥t Ä‘áº¿n má»—i Ä‘á»‰nh.
    # Ban Ä‘áº§u táº¥t cáº£ lÃ  vÃ´ cÃ¹ng (float('infinity')), trá»« Ä‘iá»ƒm xuáº¥t phÃ¡t lÃ  0.
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0

    # priority_queue (pq): HÃ ng Ä‘á»£i chá»©a cÃ¡c Ä‘á»‰nh cáº§n xÃ©t.
    # LÆ°u dÆ°á»›i dáº¡ng tuple: (khoáº£ng_cÃ¡ch, tÃªn_Ä‘á»‰nh)
    # Python sáº½ tá»± Ä‘á»™ng sáº¯p xáº¿p dá»±a trÃªn pháº§n tá»­ Ä‘áº§u tiÃªn (khoáº£ng_cÃ¡ch).
    
    priority_queue = [(0, start)]

    '''
    TRáº NG THÃI BAN Äáº¦U:
    
    Distances:  { 'A': 0,  'B': âˆ,  'C': âˆ,  'D': âˆ }
    
                  FRONT (Äáº§u hÃ ng - Min)
                    â†“
    Priority Q:   [ (0, 'A') ]
    '''

    # ---------------------------------------------------------
    # 2. VÃ’NG Láº¶P CHÃNH (MAIN LOOP)
    # ---------------------------------------------------------
    while priority_queue:
        # Láº¥y Ä‘á»‰nh cÃ³ khoáº£ng cÃ¡ch nhá» nháº¥t ra khá»i hÃ ng Ä‘á»£i
        current_dist, current_node = heapq.heappop(priority_queue)

        # Náº¿u khoáº£ng cÃ¡ch láº¥y ra lá»›n hÆ¡n khoáº£ng cÃ¡ch Ä‘Ã£ lÆ°u -> Bá» qua
        # (ÄÃ¢y lÃ  trÆ°á»ng há»£p xá»­ lÃ½ dá»¯ liá»‡u rÃ¡c/cÅ© cÃ²n sÃ³t trong hÃ ng Ä‘á»£i)
        if current_dist > distances[current_node]:
            continue

        # XÃ©t cÃ¡c hÃ ng xÃ³m cá»§a Ä‘á»‰nh hiá»‡n táº¡i
        # neighbor: tÃªn hÃ ng xÃ³m, weight: trá»ng sá»‘ cáº¡nh
        for neighbor, weight in graph[current_node].items():
            
            distance = current_dist + weight

            # -------------------------------------------------
            # 3. BÆ¯á»šC RELAXATION (Cáº¬P NHáº¬T)
            # -------------------------------------------------
            # Náº¿u tÃ¬m tháº¥y Ä‘Æ°á»ng Ä‘i ngáº¯n hÆ¡n
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                # Äáº©y hÃ ng xÃ³m vÃ o hÃ ng Ä‘á»£i vá»›i khoáº£ng cÃ¡ch má»›i
                heapq.heappush(priority_queue, (distance, neighbor))
                
                '''
                VÃ Dá»¤ MINH Há»ŒA Cáº¬P NHáº¬T:
                Khi xÃ©t tá»« C (dist=2) sang B (weight=3) -> distance = 5
                
                CÅ©: distances['B'] = 6
                Má»›i: 5 < 6  --> Cáº¬P NHáº¬T!
                
                Distances: { ... 'B': 5 ... }
                Priority Q: [ (5, 'B'), ... ]  <-- ThÃªm tráº¡ng thÃ¡i má»›i cá»§a B vÃ o
                '''

    return distances

# ---------------------------------------------------------
# CHáº Y THá»¬ Vá»šI Dá»® LIá»†U Cá»¦A CHÃšNG TA
# ---------------------------------------------------------
# Biá»ƒu diá»…n Ä‘á»“ thá»‹ báº±ng Dictionary
my_graph = {
    'A': {'B': 6, 'C': 2},
    'B': {'D': 1},
    'C': {'B': 3},
    'D': {}
}

print("Káº¿t quáº£:", dijkstra(my_graph, 'A'))

```
