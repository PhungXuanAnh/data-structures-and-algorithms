### 4. CÃ³ hÆ°á»›ng, cÃ³ trá»ng sá»‘

ÄÃ¢y lÃ  trÆ°á»ng há»£p thá»±c táº¿ nháº¥t (vÃ­ dá»¥: Báº£n Ä‘á»“ Grab/Uber, máº¡ng lÆ°á»›i chuyáº¿n bay).

Äá»ƒ dá»… hÃ¬nh dung, hÃ£y tÆ°á»Ÿng tÆ°á»£ng Ä‘Ã¢y lÃ  **Máº¡ng lÆ°á»›i vÃ© mÃ¡y bay**:

* **CÃ³ hÆ°á»›ng:** Báº¡n mua vÃ© tá»« HÃ  Ná»™i Ä‘i SÃ i GÃ²n, khÃ´ng cÃ³ nghÄ©a lÃ  vÃ© chiá»u vá» cÅ©ng tá»± Ä‘á»™ng cÃ³.
* **CÃ³ trá»ng sá»‘:** GiÃ¡ vÃ© chiá»u Ä‘i vÃ  chiá»u vá» cÃ³ thá»ƒ khÃ¡c nhau.

**Dá»¯ liá»‡u vÃ­ dá»¥:**

1. Tá»« **A** bay sang **B**: GiÃ¡ **5$**
2. Tá»« **B** bay ngÆ°á»£c vá» **A**: GiÃ¡ **8$** (Äáº¯t hÆ¡n chiá»u Ä‘i)
3. Tá»« **B** bay sang **C**: GiÃ¡ **10$**
4. **C** lÃ  Ä‘áº£o hoang, khÃ´ng cÃ³ chuyáº¿n bay nÃ o Ä‘i ra tá»« C cáº£.

---

**Quy táº¯c:**

* HÃ ng (Row) lÃ  nÆ¡i **Xuáº¥t phÃ¡t**.
* Cá»™t (Col) lÃ  nÆ¡i **Äáº¿n**.
* Ã” giao nhau chá»©a **GiÃ¡ tiá»n (Trá»ng sá»‘)**.

```text
      (Äáº¾N)
      A    B    C
    +----+----+----+
(Tá»ª)|    |    |    |

    +----+----+----+
(Tá»ª)|    |    |    |
  B | 8  | 0  | 10 |  <- Tá»« B vá» A tá»‘n 8$. Tá»« B Ä‘áº¿n C tá»‘n 10$
    +----+----+----+

    +----+----+----+

```

**Äiá»ƒm cáº§n chÃº Ã½:**

* NhÃ¬n vÃ o Ã´ `(A, B)` lÃ  **5**, nhÆ°ng nhÃ¬n vÃ o Ã´ Ä‘á»‘i xá»©ng `(B, A)` láº¡i lÃ  **8**.
*  Ma tráº­n **khÃ´ng Ä‘á»‘i xá»©ng** qua Ä‘Æ°á»ng chÃ©o chÃ­nh.

### 5. Code python

DÃ¹ng cho Ä‘á»“ thá»‹ dÃ y Ä‘áº·c (nhiá»u Ä‘Æ°á»ng ná»‘i).


````python

class GraphMatrix:
    def __init__(self, size):
        # size: Sá»‘ lÆ°á»£ng Ä‘á»‰nh (vÃ­ dá»¥ 4 Ä‘á»‰nh A,B,C,D)
        self.size = size
        
        # Táº¡o má»™t báº£ng lÆ°á»›i vuÃ´ng (size x size) toÃ n sá»‘ 0
        # VÃ­ dá»¥ size=4 thÃ¬ táº¡o 4 hÃ ng, má»—i hÃ ng 4 sá»‘ 0
        self.matrix = []
        for i in range(size):
            self.matrix.append([0] * size)

    def add_edge(self, u, v, weight=1, directed=False):
        # u: Äá»‰nh nguá»“n (dáº¡ng sá»‘, vÃ­ dá»¥ 0 lÃ  A)
        # v: Äá»‰nh Ä‘Ã­ch (dáº¡ng sá»‘, vÃ­ dá»¥ 1 lÃ  B)
        # weight: Trá»ng sá»‘ (máº·c Ä‘á»‹nh lÃ  1)
        # directed: CÃ³ hÆ°á»›ng hay khÃ´ng?

        # GÃ¡n trá»ng sá»‘ vÃ o Ã´ giao nhau giá»¯a hÃ ng u vÃ  cá»™t v
        self.matrix[u][v] = weight

        # Náº¿u lÃ  vÃ´ hÆ°á»›ng (2 chiá»u), ta pháº£i gÃ¡n ngÆ°á»£c láº¡i tá»« v vá» u
        if not directed:
            self.matrix[v][u] = weight

    def print_matrix(self):
        print("--- Ma tráº­n ká» ---")
        vertex_names = ['A', 'B', 'C', 'D']  # TÃªn cÃ¡c Ä‘á»‰nh
        
        for i, row in enumerate(self.matrix):
            # In hÃ ng dá»¯ liá»‡u
            print(row, end="")
            
            # Táº¡o chÃº thÃ­ch cho hÃ ng
            print(f"  <- HÃ ng {i} ({vertex_names[i]}): ", end="")
            
            # Liá»‡t kÃª cÃ¡c Ä‘Æ°á»ng Ä‘i tá»« Ä‘á»‰nh nÃ y
            connections = []
            for j, weight in enumerate(row):
                if weight != 0:
                    connections.append(f"Ä‘áº¿n {vertex_names[j]}(cá»™t {j}) giÃ¡ {weight}")
            
            if connections:
                print(", ".join(connections))
            else:
                print("ToÃ n 0 (KhÃ´ng Ä‘i Ä‘Ã¢u Ä‘Æ°á»£c)")

# === CHáº Y THá»¬ (MÃ´ phá»ng láº¡i HÃ¬nh 2 - CÃ³ hÆ°á»›ng, CÃ³ trá»ng sá»‘) ===
# Giáº£ sá»­: A=0, B=1, C=2, D=3
g = GraphMatrix(4) 

# A(0) -> B(1) trá»ng sá»‘ 3
g.add_edge(0, 1, weight=3, directed=True)
# A(0) -> C(2) trá»ng sá»‘ 2
g.add_edge(0, 2, weight=2, directed=True)
# C(2) -> B(1) trá»ng sá»‘ 1
g.add_edge(2, 1, weight=1, directed=True)
# D(3) -> A(0) trá»ng sá»‘ 4
g.add_edge(3, 0, weight=4, directed=True)

g.print_matrix()
````


## PHáº¦N 2: DANH SÃCH Ká»€ (ADJACENCY LIST)

**TÆ°á»Ÿng tÆ°á»£ng:** Danh báº¡ Ä‘iá»‡n thoáº¡i.

* **Quy táº¯c:** Má»—i Ä‘á»‰nh cÃ³ má»™t danh sÃ¡ch riÃªng, chá»‰ ghi tÃªn nhá»¯ng "ngÆ°á»i hÃ ng xÃ³m" mÃ  nÃ³ cÃ³ thá»ƒ Ä‘i trá»±c tiáº¿p Ä‘áº¿n.
* **Æ¯u Ä‘iá»ƒm:** Tiáº¿t kiá»‡m bá»™ nhá»›, chá»‰ lÆ°u nhá»¯ng gÃ¬ cÃ³ káº¿t ná»‘i. ÄÃ¢y lÃ  cÃ¡ch dÃ¹ng phá»• biáº¿n nháº¥t trong thá»±c táº¿.



### 1. KhÃ´ng trá»ng sá»‘, VÃ´ hÆ°á»›ng

*MÃ´ táº£:* A ná»‘i B, B ná»‘i C, A ná»‘i C.

```text
[Äá»‰nh] -> [Danh sÃ¡ch hÃ ng xÃ³m]
------------------------------
  A    -> [ B, C ]  (A quen B vÃ  C)
  B    -> [ A, C ]  (B quen A vÃ  C)
  C    -> [ A, B ]  (C quen A vÃ  B)

```



### 2. CÃ³ trá»ng sá»‘ (Giáº£ sá»­ vÃ´ hÆ°á»›ng)

*MÃ´ táº£:* A-B (5), B-C (10), A-C (2).
*Quy táº¯c:* Trong danh sÃ¡ch, thay vÃ¬ chá»‰ ghi tÃªn, ta ghi má»™t cáº·p `(TÃªn hÃ ng xÃ³m, Trá»ng sá»‘)`.

```text
[Äá»‰nh] -> [Danh sÃ¡ch (HÃ ng xÃ³m, Sá»‘ km)]
---------------------------------------
  A    -> [ (B, 5), (C, 2) ]  (A Ä‘áº¿n B tá»‘n 5, Ä‘áº¿n C tá»‘n 2)
  B    -> [ (A, 5), (C, 10)]  (B vá» A tá»‘n 5, Ä‘áº¿n C tá»‘n 10)
  C    -> [ (A, 2), (B, 10)]  (C vá» A tá»‘n 2, vá» B tá»‘n 10)

```



### 3. CÃ³ hÆ°á»›ng (Giáº£ sá»­ khÃ´ng trá»ng sá»‘)

*MÃ´ táº£:* Äi tá»« A  B, vÃ  tá»« B  C.

```text
[Äá»‰nh] -> [Danh sÃ¡ch cÃ¡c nÆ¡i ÄI ÄÆ¯á»¢C Ä‘áº¿n]
-----------------------------------------
  A    -> [ B ]    (A chá»‰ Ä‘i Ä‘Æ°á»£c Ä‘áº¿n B)
  B    -> [ C ]    (B chá»‰ Ä‘i Ä‘Æ°á»£c Ä‘áº¿n C)
  C    -> [ ]      (C khÃ´ng Ä‘i Ä‘Æ°á»£c Ä‘Ã¢u háº¿t - Danh sÃ¡ch rá»—ng)

```

### 4. CÃ³ hÆ°á»›ng cÃ³ trá»ng sá»‘


**Quy táº¯c:**

* Chá»‰ lÆ°u nhá»¯ng chuyáº¿n bay **cÃ³ tháº­t**.
* Má»—i má»¥c trong danh sÃ¡ch pháº£i káº¹p nÃ¡ch thÃªm giÃ¡ tiá»n: `(NÆ¡i Ä‘áº¿n, GiÃ¡ tiá»n)`.

```text
[SÃ¢n bay] -> [Danh sÃ¡ch cÃ¡c chuyáº¿n Ä‘i Ä‘Æ°á»£c: (NÆ¡i Ä‘áº¿n, GiÃ¡)]
----------------------------------------------------------
    A     -> [ (B, 5) ]           
             (Tá»« A chá»‰ bay Ä‘Æ°á»£c Ä‘áº¿n B vá»›i giÃ¡ 5)

    B     -> [ (A, 8), (C, 10) ]  
             (Tá»« B bay Ä‘Æ°á»£c vá» A giÃ¡ 8, vÃ  sang C giÃ¡ 10)

    C     -> [ ]                  
             (C khÃ´ng cÃ³ chuyáº¿n bay nÃ o cáº¥t cÃ¡nh - Danh sÃ¡ch rá»—ng)

```

---

### 5. code python

````python

class GraphList:
    def __init__(self):
        # Táº¡o má»™t tá»« Ä‘iá»ƒn rá»—ng Ä‘á»ƒ chá»©a danh báº¡
        # Cáº¥u trÃºc sáº½ lÃ : { "A": [], "B": [] ... }
        self.adj_list = {}

    def add_node(self, node):
        # Náº¿u Ä‘á»‰nh nÃ y chÆ°a cÃ³ trong danh báº¡ thÃ¬ táº¡o má»›i vá»›i list rá»—ng
        if node not in self.adj_list:
            self.adj_list[node] = []

    def add_edge(self, u, v, weight=1, directed=False):
        # u: TÃªn Ä‘á»‰nh Ä‘i ("A"), v: TÃªn Ä‘á»‰nh Ä‘áº¿n ("B")
        
        # BÆ°á»›c 1: Äáº£m báº£o 2 Ä‘á»‰nh nÃ y Ä‘Ã£ cÃ³ trong danh sÃ¡ch quáº£n lÃ½
        self.add_node(u)
        self.add_node(v)

        # BÆ°á»›c 2: ThÃªm Ä‘Æ°á»ng Ä‘i tá»« u -> v
        # Ta lÆ°u má»™t bá»™ (tuple) gá»“m (TÃªn Ä‘á»‰nh Ä‘áº¿n, Trá»ng sá»‘)
        # TÆ°Æ¡ng tá»± cÃ¡i Ã´ vuÃ´ng [1, 3] trong hÃ¬nh váº½ cá»§a báº¡n
        self.adj_list[u].append((v, weight))

        # BÆ°á»›c 3: Náº¿u lÃ  vÃ´ hÆ°á»›ng (khÃ´ng cÃ³ mÅ©i tÃªn), thÃªm chiá»u ngÆ°á»£c láº¡i
        if not directed:
            self.adj_list[v].append((u, weight))

    def print_graph(self):
        print("--- Danh sÃ¡ch ká» ---")
        # Duyá»‡t qua tá»«ng Ä‘á»‰nh vÃ  in danh sÃ¡ch hÃ ng xÃ³m
        for node, neighbors in self.adj_list.items():
            print(f"Äá»‰nh {node} Ä‘i tá»›i: {neighbors}")

# === CHáº Y THá»¬ (MÃ´ phá»ng láº¡i HÃ¬nh 2 - CÃ³ hÆ°á»›ng, CÃ³ trá»ng sá»‘) ===
g_list = GraphList()

# A -> B (3)
g_list.add_edge("A", "B", weight=3, directed=True)
# A -> C (2)
g_list.add_edge("A", "C", weight=2, directed=True)
# C -> B (1) (LÆ°u Ã½: Pháº£i gá»i add_edge Ä‘Ãºng chiá»u)
g_list.add_edge("C", "B", weight=1, directed=True)
# D -> A (4)
g_list.add_edge("D", "A", weight=4, directed=True)

g_list.print_graph()
````


## PHáº¦N 3: PSEUDO-CODE CHO DANH SÃCH Ká»€

### 5. Python code
Giáº£ sá»­ chÃºng ta cÃ³ má»™t sÆ¡ Ä‘á»“ Ä‘Æ¡n giáº£n dÆ°á»›i dáº¡ng Python Dictionary (Danh sÃ¡ch ká»):

````python
graph = {
    'A': ['B', 'C'],   # Tá»« A Ä‘i Ä‘Æ°á»£c sang B vÃ  C
    'B': ['D', 'E'],   # Tá»« B Ä‘i Ä‘Æ°á»£c sang D vÃ  E
    'C': ['F'],        # Tá»« C Ä‘i Ä‘Æ°á»£c sang F
    'D': [], 'E': [], 'F': [] # NgÃµ cá»¥t
}

def dfs_demo(graph, start_node):
    visited = []            # Sá»• tay
    stack = [start_node]    # NgÄƒn xáº¿p (dÃ¹ng list lÃ m stack)

    while stack:
        # 0. In ra tráº¡ng thÃ¡i hiá»‡n táº¡i
        print(f"NgÄƒn xáº¿p hiá»‡n táº¡i: {stack}")
        
        # 1. Láº¥y ngÆ°á»i TRÃŠN CÃ™NG (vá»«a má»›i vÃ o) ra
        node = stack.pop()  
        
        if node not in visited:
            visited.append(node)
            print(f"Äang thÄƒm: {node}")

            # 2. Äáº©y hÃ ng xÃ³m vÃ o ngÄƒn xáº¿p (Äáº¢O NGÆ¯á»¢C Ä‘á»ƒ Æ°u tiÃªn ngÆ°á»i Ä‘á»©ng Ä‘áº§u danh sÃ¡ch gá»‘c,
            # cho Ä‘Ãºng vá»›i giáº£ sá»­ trong pháº§n # Cháº¡y thuáº­t toÃ¡n tá»«ng bÆ°á»›c)
            for neighbor in reversed(graph[node]):
                stack.append(neighbor)
            
    print("Káº¿t thÃºc DFS. CÃ¡c Ä‘á»‰nh Ä‘Ã£ thÄƒm:", visited)
    
# Cháº¡y thá»­ DFS tá»« Ä‘á»‰nh 'A'
dfs_demo(graph, 'A')
````
