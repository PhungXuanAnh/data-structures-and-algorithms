## 2. Váº¥n Ä‘á» Chu trÃ¬nh Ã¢m (Negative Cycle) ğŸ•³ï¸

Chu trÃ¬nh Ã¢m lÃ  má»™t "lá»— há»•ng thá»i gian". VÃ­ dá»¥: `A -> B -> C -> A` cÃ³ tá»•ng trá»ng sá»‘ lÃ  `-1`.
Cá»© Ä‘i má»™t vÃ²ng, báº¡n láº¡i giáº£m Ä‘Æ°á»£c 1 Ä‘Æ¡n vá»‹ chi phÃ­. Báº¡n Ä‘i 1 tá»· vÃ²ng, chi phÃ­ giáº£m Ä‘i 1 tá»·. KhÃ´ng bao giá» cÃ³ Ä‘iá»ƒm dá»«ng nhá» nháº¥t.

**CÃ¡ch phÃ¡t hiá»‡n:** Sau khi Ä‘Ã£ láº·p Ä‘á»§ `V-1` láº§n (vá» lÃ½ thuyáº¿t lÃ  Ä‘Ã£ xong viá»‡c), ta **thá»­ cáº­p nháº­t thÃªm má»™t láº§n ná»¯a** (Láº§n thá»© `V`).

* Náº¿u má»i thá»© váº«n yÃªn áº¯ng: Äá»“ thá»‹ an toÃ n.
* Náº¿u váº«n cÃ³ má»™t cáº¡nh nÃ o Ä‘Ã³ **giáº£m Ä‘Æ°á»£c chi phÃ­ tiáº¿p**: CHáº®C CHáº®N CÃ“ CHU TRÃŒNH Ã‚M. (VÃ¬ Ä‘Æ°á»ng Ä‘i Ä‘Æ¡n giáº£n khÃ´ng thá»ƒ dÃ i quÃ¡ `V-1` cáº¡nh, náº¿u cáº­p nháº­t tiáº¿p nghÄ©a lÃ  nÃ³ Ä‘ang cháº¡y vÃ²ng trÃ²n).

**Logic phÃ¡t hiá»‡n Chu trÃ¬nh Ã¢m**: Theo cÃ¡ch trÃªn, Ä‘á»ƒ xÃ¡c Ä‘á»‹nh chu trÃ¬nh Ã¢m, ta thá»±c hiá»‡n thÃªm má»™t vÃ²ng láº·p thá»© `V` (sau khi Ä‘Ã£ cháº¡y xong `V-1` vÃ²ng chÃ­nh). Ta duyá»‡t qua táº¥t cáº£ cÃ¡c cáº¡nh cÃ³ trá»ng sá»‘ :
* **Kiá»ƒm tra:** `dist[u] + w < dist[v]`
* **Ã nghÄ©a:** Náº¿u Ä‘iá»u kiá»‡n nÃ y Ä‘Ãºng, tá»©c lÃ  ta váº«n tÃ¬m Ä‘Æ°á»£c má»™t con Ä‘Æ°á»ng ngáº¯n hÆ¡n má»©c "tá»‘i Æ°u" lÃ½ thuyáº¿t.
* **Káº¿t luáº­n:** Äá»“ thá»‹ chá»©a chu trÃ¬nh Ã¢m (Negative Cycle), vÃ¬ chá»‰ cÃ³ viá»‡c Ä‘i vÃ²ng trÃ²n qua cÃ¡c cáº¡nh Ã¢m má»›i cÃ³ thá»ƒ lÃ m giáº£m chi phÃ­ thÃªm ná»¯a sau khi Ä‘Ã£ duyá»‡t Ä‘á»§ sá»‘ cáº¡nh tá»‘i Ä‘a cá»§a má»™t Ä‘Æ°á»ng Ä‘i Ä‘Æ¡n.

**Xem pháº§n minh há»a xÃ¡c Ä‘á»‹nh chu trÃ¬nh Ã¢m á»Ÿ pháº§n implement distances sá»­ dá»¥ng dict thay vÃ¬ list**

## 3. Code Python thá»±c chiáº¿n & Cháº¡y báº±ng cÆ¡m ğŸš

DÆ°á»›i Ä‘Ã¢y lÃ  Ä‘oáº¡n code Python Ä‘áº§y Ä‘á»§, bao gá»“m viá»‡c phÃ¡t hiá»‡n chu trÃ¬nh Ã¢m vÃ  cháº¡y vÃ­ dá»¥ "HoÃ n tiá»n":

````python
# Äáº¡i diá»‡n giÃ¡ trá»‹ VÃ´ cÃ¹ng
INF = float('inf')

class Graph:
    def __init__(self, vertices):
        self.V = vertices  # Sá»‘ lÆ°á»£ng Ä‘á»‰nh
        self.edges = []    # Danh sÃ¡ch lÆ°u cÃ¡c cáº¡nh: [v1, v2, weight]

    def add_edge(self, v1, v2, w):
        self.edges.append([v1, v2, w])

    def bellman_ford(self, start_node_index):
        # 1. KHá»I Táº O
        # dist: Máº£ng lÆ°u khoáº£ng cÃ¡ch tá»« start_node Ä‘áº¿n cÃ¡c Ä‘á»‰nh khÃ¡c
        dist = [INF] * self.V
        dist[start_node_index] = 0

        # ========================================================
        # 2. VÃ’NG Láº¶P CHÃNH (RELAXATION) - Cháº¡y V-1 láº§n
        # ========================================================
        print("\n--- Báº®T Äáº¦U CHáº Y THUáº¬T TOÃN ---")
        
        for i in range(self.V - 1):
            print(f"ğŸ”„ VÃ²ng láº·p thá»© {i+1}:")
            
            # Duyá»‡t qua tá»«ng cáº¡nh trong Ä‘á»“ thá»‹
            for v1, v2, w in self.edges:
                # Náº¿u Ä‘Ã£ Ä‘áº¿n Ä‘Æ°á»£c v1 (khÃ¡c INF) VÃ€ Ä‘i qua v1 ráº» hÆ¡n cÃ¡ch cÅ©
                if dist[v1] != INF and dist[v1] + w < dist[v2]:
                    
                    print(f"   âš¡ Cáº­p nháº­t: Tá»« Ä‘á»‰nh {v1} sang {v2}")
                    print(f"      CÅ©: {dist[v2]} | Má»›i: {dist[v1]} + ({w}) = {dist[v1] + w}")
                    
                    dist[v2] = dist[v1] + w
            
            print(f"   ğŸ“Š Tráº¡ng thÃ¡i dist: {dist}")

        # ========================================================
        # 3. KIá»‚M TRA CHU TRÃŒNH Ã‚M (Láº§n thá»© V)
        # ========================================================
        print("\nğŸ•µï¸ Kiá»ƒm tra chu trÃ¬nh Ã¢m...")
        for v1, v2, w in self.edges:
            if dist[v1] != INF and dist[v1] + w < dist[v2]:
                print("âš ï¸ Cáº¢NH BÃO: PhÃ¡t hiá»‡n chu trÃ¬nh Ã¢m!")
                return None

        print("\nâœ… ÄÃ£ tÃ¬m tháº¥y Ä‘Æ°á»ng Ä‘i ngáº¯n nháº¥t!")
        return dist

# --- CHáº Y THá»¬ (DRIVER CODE) ---

# Táº¡o Ä‘á»“ thá»‹ cÃ³ 3 Ä‘á»‰nh: 0(A), 1(B), 2(C)
g = Graph(3)

# ThÃªm cÃ¡c cáº¡nh nhÆ° vÃ­ dá»¥ "HoÃ n tiá»n"
# A -> B (2)
g.add_edge(0, 1, 2)

# A -> C (5)
g.add_edge(0, 2, 5)

# C -> B (-10)  <-- Cáº¡m báº«y
g.add_edge(2, 1, -10)

# Cháº¡y thuáº­t toÃ¡n tá»« Ä‘á»‰nh 0 (A)
ket_qua = g.bellman_ford(0)

print("\nKáº¾T QUáº¢ CUá»I CÃ™NG (Tá»« A):")
if ket_qua:
    mapping = {0: 'A', 1: 'B', 2: 'C'}
    for i in range(3):
        print(f"Äáº¿n {mapping[i]}: {ket_qua[i]}")
````

### 5. Code Python hoÃ n chá»‰nh (DÃ¹ng Dict + A, B, C + Minh há»a Chu trÃ¬nh Ã¢m)

ÄÃ¢y lÃ  phiÃªn báº£n code dá»… Ä‘á»c nháº¥t, há»— trá»£ tÃªn Ä‘á»‰nh A, B, C vÃ  bao gá»“m vÃ­ dá»¥ vá» **Chu trÃ¬nh Ã¢m thá»±c sá»±**.

````python
INF = float('inf')

class Graph:
    def __init__(self):
        self.edges = []    # LÆ°u cáº¡nh: {'v1': 'A', 'v2': 'B', 'w': 1}
        self.nodes = set() # LÆ°u táº­p há»£p tÃªn cÃ¡c Ä‘á»‰nh: {'A', 'B', 'C'}

    def add_edge(self, v1, v2, w):
        self.edges.append([v1, v2, w])
        self.nodes.add(v1)
        self.nodes.add(v2)

    def bellman_ford(self, start_node):
        # 1. KHá»I Táº O (DÃ¹ng Dict nhÆ° báº¡n yÃªu cáº§u)
        dist = {node: INF for node in self.nodes}
        dist[start_node] = 0
        
        num_vertices = len(self.nodes)

        print(f"\nğŸš€ Báº®T Äáº¦U tá»« Ä‘á»‰nh [{start_node}] vá»›i {num_vertices} Ä‘á»‰nh")
        print("-" * 40)

        # 2. VÃ’NG Láº¶P CHÃNH (V-1 láº§n)
        for i in range(num_vertices - 1):
            changed = False # Cá» kiá»ƒm tra xem vÃ²ng nÃ y cÃ³ gÃ¬ thay Ä‘á»•i ko
            
            print(f"ğŸ”„ VÃ²ng {i+1}:")
            for v1, v2, w in self.edges:
                # Náº¿u Ä‘Æ°á»ng má»›i tá»‘t hÆ¡n Ä‘Æ°á»ng cÅ©
                if dist[v1] != INF and dist[v1] + w < dist[v2]:
                    print(f"   âš¡ Cáº­p nháº­t [{v1} -> {v2}]:")
                    print(f"      CÅ©: {dist[v2]} | Má»›i: {dist[v1]} + ({w}) = {dist[v1] + w}")
                    
                    dist[v2] = dist[v1] + w
                    changed = True
            
            # In tráº¡ng thÃ¡i sau má»—i vÃ²ng
            # Sáº¯p xáº¿p key Ä‘á»ƒ in ra cho Ä‘áº¹p A, B, C
            sorted_dist = dict(sorted(dist.items()))
            print(f"   ğŸ“Š Dist: {sorted_dist}")
            
            # Tá»‘i Æ°u: Náº¿u vÃ²ng nÃ y khÃ´ng cÃ³ gÃ¬ thay Ä‘á»•i thÃ¬ dá»«ng sá»›m luÃ´n
            if not changed:
                print("   ğŸ’¤ KhÃ´ng cÃ³ thay Ä‘á»•i nÃ o, dá»«ng vÃ²ng láº·p sá»›m.")
                break

        # 3. KIá»‚M TRA CHU TRÃŒNH Ã‚M (Láº§n thá»© V)
        print("\nğŸ•µï¸ Kiá»ƒm tra 'Lá»— Ä‘en' (Chu trÃ¬nh Ã¢m)...")
        for v1, v2, w in self.edges:
            if dist[v1] != INF and dist[v1] + w < dist[v2]:
                print(f"âš ï¸ Cáº¢NH BÃO: PhÃ¡t hiá»‡n chu trÃ¬nh Ã¢m táº¡i cáº¡nh [{v1} -> {v2}]!")
                print(f"   VÃ¬ {dist[v1]} + ({w}) < {dist[v2]}")
                print("   ğŸ˜± CÃ ng cháº¡y cÃ ng giáº£m -> KhÃ´ng cÃ³ Ä‘Ã¡p Ã¡n cuá»‘i cÃ¹ng.")
                return None # KhÃ´ng cÃ³ káº¿t quáº£ vÃ¬ chu trÃ¬nh Ã¢m, khÃ´ng cÃ³ Ä‘Æ°á»ng Ä‘i ngáº¯n nháº¥t Ä‘áº¿n cÃ¡c Ä‘á»‰nh

        print("\nâœ… Äá»“ thá»‹ an toÃ n. Káº¿t quáº£ cuá»‘i cÃ¹ng:")
        return dist

# --- TEST CASE: CHU TRÃŒNH Ã‚M ---
# HÃ£y xem vÃ­ dá»¥ dÆ°á»›i Ä‘Ã¢y Ä‘á»ƒ hiá»ƒu "Lá»— Ä‘en"
# A -> B (1)
# B -> C (-5)
# C -> B (2)  <-- VÃ²ng láº·p B <-> C: Ä‘i 1 vÃ²ng tá»‘n -5 + 2 = -3 (LÃƒI 3 Äá»’NG)

print("--- VÃ Dá»¤ CHU TRÃŒNH Ã‚M --- xem minh há»a 'Lá»— Ä‘en' bÃªn dÆ°á»›i ---")
g_negative = Graph()
g_negative.add_edge('A', 'B', 1)
g_negative.add_edge('B', 'C', -5)
g_negative.add_edge('C', 'B', 2) 

ket_qua = g_negative.bellman_ford('A')
````
